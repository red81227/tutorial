```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Ë´ãÁπ™Âá∫ùëì(ùë•)=sin‚Å°(ùë•), ‚àí2‚â§ùë•‚â§2

```python
x = np.arange(-2.0,2.0,0.01) 
y_sin = np.sin(2 * np.pi * x)
y_cos = np.cos(2 * np.pi * x)

#ÂÆöÁæ©‰∏ÄÂÄãÂúñÂÉèÁ™óÂè£
fig = plt.figure()

#ÂÆöÁæ©x,yÂíåÂúñÁöÑÊ®£Âºè
plt.plot(x, y_sin, '.')
```

## Ë´ãÁπ™Âá∫ùëì(ùë•)=sin‚Å°(ùë•), ‚àí2‚â§ùë•‚â§2Âèäùëì(ùë•)=ùëêùëúùë†‚Å°(ùë•), ‚àí2‚â§ùë•‚â§2,
## Âú®Âêå‰∏ÄÂºµÂúñÁ¥ô‰∏ä

```python
#Áπ™Ë£ΩsinËàácosÁöÑÂúñ
fig = plt.figure() 
plt.plot(x, y_sin, '.', x, y_cos, '--')
```

# [Á∑¥Áøí]Ë´ãÂèÉÁÖß‰∏äÈù¢‰æãÂ≠ê, Áπ™Âá∫ùëì(ùë•)= ùë•^2, ‚àí2‚â§ùë•‚â§2

```python
x = np.arange(-2.0, 2.0, 0.01) 
y = x**2 
plt.plot(x, y) 
plt.xlabel("x") 
plt.ylabel("y") 
plt.title("$f(x) = x^2$") 
plt.grid(True) 
plt.show() 
```

## Áõ¥ÊñπÂúñ(Histogram)
Ôºç‰∫åÁ∂≠Áµ±Ë®àÂúñË°®ÔºåÊ©´Ëª∏ÊòØÈ°ûÂà•ÊàñÊòØÊï∏ÂÄºÈñìÁöÑÁØÑÂúçÔºåÁ∏±Ëª∏ÊòØ‰∫ã‰ª∂ÁôºÁîüÁöÑÈ†ªÁéá„ÄÇ

```python
fig = plt.figure()
x_hist = np.random.randn(1000)
plt.hist(x_hist, bins=20, color='SteelBlue', edgecolor="LightSteelBlue") 
```

## ÂúìÈ§ÖÂúñ(Pie Chart)
ÔºçÂúìÂΩ¢Áµ±Ë®àÂúñË°®ÔºåÊØèÂÄãÊâáÂçÄÁöÑÂºßÈï∑Ë°®Á§∫ÁöÑÊï∏ÈáèÁöÑÊØî‰æã„ÄÇ

```python
fig = plt.figure()
labels = ["day1","day2","day3","day3","day5","day6"]
word_size = [2000, 3250, 4600 , 2300, 3600, 5600]
explode = [0 , 0, 0.1, 0, 0, 0]  #0Ââá‰∏çÁ™ÅÂá∏Âá∫ÔºåÂÄºË∂äÂ§ßÂâáÂá∏Âá∫Ë∂äÂ§ß
plt.pie(word_size, explode= explode, labels = labels, autopct= "%3.1f%%")
```

## Áõ¥Ê¢ùÂúñ(Bar Chart)

```python
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')
y_pos = np.arange(len(bars))
 
# Create bars
plt.bar(y_pos, height)
 
# Create names on the x-axis
plt.xticks(y_pos, bars)
plt.show()
```

## ÁÆ±ÂûãÂúñ(Box Plot)

```python
np.random.seed(1234)

# fake up some data
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))

fig1, ax1 = plt.subplots()
ax1.set_title('Basic Plot')
ax1.boxplot(data)
```
