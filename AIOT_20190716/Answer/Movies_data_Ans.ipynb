```python
import pandas as pd
import numpy as np
```

## Movies
star_rating：星等 <br>
title：片名<br>
content_rating：內容分級（G：普遍級, PG：輔導級, PG-13：輔導13歲級, R：限制17歲級, NC-17：限制級）<br>
genre：影片類型<br>
duration：片長<br>
actors_list：演員陣容

```python
movies = pd.read_csv("http://bit.ly/imdbratings")
movies.shape
```

```python
movies.head()
```

### 根據片名排序

```python
movies.title.sort_values(ascending = True)
```

### 先依照內容分級再依照片長排序

```python
movies.sort_values(['content_rating','duration'])
```

### 找出片長大於200分鐘且為冒險的影片

```python
movies[(movies.duration >= 200) & (movies.genre == 'Adventure')]
```

### 找出為恐怖或冒險或動作的影片

```python
movies[movies.genre.isin(['Crime', 'Adventure', 'Action'])]
```
