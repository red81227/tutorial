#### [練習] 將 Temperature_Taipei 中的所有溫度都轉成華氏
Temperature_Taipei = [22, 24, 19, 19, 30, 21, 25]

公式： Fahrenheit=Celsius×(9/5)+32

```python
Temperature_array = np.array([22, 24, 19, 19, 30, 21, 25])

Temperature_array_Fahrenheit = (Temperature_array) * ( 9 / 5 ) + 32
print(Temperature_array_Fahrenheit)
```

#### [練習] 利用array, 寫出一張九九乘法表

```python
multiplication = np.empty( (9 , 9) ) 

for i in range(1,10): 
    for j in range(1,10): 
        multiplication[(i-1),(j-1)]= i*j
print(multiplication)
```
